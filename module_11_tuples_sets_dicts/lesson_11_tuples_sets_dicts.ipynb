# Module 11: Tuples, Sets & Dictionaries

Python has four primary collection types. This module covers the remaining three after lists.

## 11.1 Tuples — Immutable Sequences

Tuples are like lists, but **immutable** (cannot be changed after creation). Use them for data that should not change.

```python
# Creating tuples
coords = (3.5, 7.2)          # typical
rgb    = (255, 128, 0)
singleton = (42,)             # single item needs trailing comma!
nested = ((1, 2), (3, 4))

# Access (same as list)
print(coords[0])   # 3.5
print(rgb[-1])     # 0
print(len(rgb))    # 3

# Immutable — this would fail:
# coords[0] = 99  # TypeError!

# Packing and unpacking
point = 10, 20        # packing (parens optional)
x, y = point          # unpacking
print(x, y)           # 10 20

a, b, *rest = (1, 2, 3, 4, 5)
print(a, b, rest)     # 1 2 [3, 4, 5]
```

### When to use tuples vs lists

| Tuples | Lists |
|--------|-------|
| Fixed data (coordinates, RGB, DB rows) | Data that changes |
| Function return values | Working with a collection |
| Dictionary keys (they're hashable) | Not usable as dict keys |
| Slightly faster than lists | Slightly slower |

## 11.2 Sets — Unique Unordered Collections

A set is an **unordered collection of unique** values. Duplicate items are automatically removed.

```python
# Creating sets
colors = {"red", "green", "blue"}
print(colors)

nums   = {1, 2, 3, 4, 5}
print(nums)

# To create an empty set, you have to use the set() function
empty  = set()
print(empty)

# Duplicates are removed
dupes = {1, 2, 2, 3, 3, 3}
print(dupes)

# Basic operations
colors.add("purple")
print(colors)

colors.remove("red")  # raises KeyError if not found
print(colors)

colors.discard("yellow")   # no error if not found
print(colors)

# .clear() method removes all elements from a set
# colors.clear()
# print(colors) # will print set(), because it is empty

# Membership testing
print("red" in colors)    # True
print("purple" in colors) # False
```

```python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7}

# Union operator '|' returns new set with all elements from both sets
a | b
print("Union:       ", a | b)

# Intersection operator '&' returns new set with only common elements
a & b
print("Intersection:", a & b)

# Difference operator '-' returns new set with elements of first set that are NOT in other sets
a - b
print("Difference:  ", a - b)

# Symmetric operator '^' returns new set with elements that are either in {a} or {b}, but not both
a ^ b
print("Sym diff:    ", a ^ b)

# We can use the compound assignment operator to assign the resulting set to the first set in the expression
a -= b # Finds difference between sets, and updates {a} with that result


# .issubset() and .issuperset() methods check if a set is a subset or superset of another set
# Check is {1,2} is a subset of {1,2,3}: all elements of {1,2} are in {1,2,3}, so the output is True
print({1,2}.issubset({1,2,3}))

# Check if {1,2,3} is a superset of {1,2}: {1,2,3} has all the elements of {1,2}, so the output is True
print({1,2,3}.issuperset({1,2}))

# .isdisjoint() method checks if two sets are disjoint, meaning they don't have any elements in common
print(a.isdisjoint(b)) # False, because no element in {a} is in {b}
```

## 11.3 Dictionaries — Key-Value Pairs

Dictionaries map **keys** to **values**. Keys must be unique and immutable (strings, numbers, tuples). Values can be anything.

```python
# Creating a dict
person = {
    "name": "Alice",  # key1: value1
    "age": 30,        # key2: value2
    "city": "London", # key 3: value3
    "hobbies": ["reading", "coding"] # key4: value4
}

# We can also use the dict() constructor, which builds the dictionary from a sequence of key-value pairs
person = dict([("name:", "Alice"), ("age:", 30), ("city:", "London"), ("hobbies:", ["coding", "reading"])])

# Access by key using bracket notation: dictionary[key]
person["name"]     # Alice
person["hobbies"]  # ['reading', 'coding']

# .get() method retrieves the value associated with a key, and you can set a default value so you don't get an error if key doesn't exist
# dictionary.get(key, default)
person.get("age")        # 30
person.get("salary")     # will return None, because there is no default set
person.get("salary", 0)  # will return 0, because default = 0

# Add or modify
person["email"] = "alice@example.com"
person["age"] = 31
```

```python
person = {"a": 1, "b": 2, "c": 3, "d": 4}

# .pop() method removes the key-value pair with the key you specify as 1st argument, and returns its value
val  = person.pop("b")       # remove & return value
print(val, person)

del person["c"]              # delete entry
print(person)

# .popitem() method removes the last inserted item
last = person.popitem()
print(last, person)
```

```python
# Other methods
d = {"a": 1, "b": 2}

# .keys() and .values() methods return a view object with all the keys and values in the dictionary
d.keys()      # dict_keys(['a', 'b'])
d.values()    # dict_values([1, 2])

# .items() method returns a view object with all the key-value pairs in the dictionary
d.items()     # dict_items([('a', 1), ('b', 2)])

# .clear() method removes all the key-value pairs from the dictionary
d.clear()

print(len(d))          # 2
print("a" in d)        # True ← checks if the key "a" is in d
print(1 in d.values()) # True ← checks if the value 1 is in d

# Merge dictionaries
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}
merged = d1 | d2
print(merged)

# .update() method updates key-value pairs with key-value pairs from another dictionary. Common keys -> values are overwritten
d1.update({"b": 99, "e": 5})
print(d1)
```

```python
# Looping over a dictionary
data = {"x": 10, "y": 20, "z": 30}

for key in data:                      # loop variable is 'key', and you loop over the dictionary 'data' (default is keys)
    print(key)

for key in data.keys():               # same as above, but instead we use the .keys() method
    print(key)

for val in data.values():             # loop variable is 'val', and you loop over the dictionary 'data'
    print(val)

for key, val in data.items():         # loop variable is both 'key' and 'val', loop over entire dictionary, keeps 'key' and 'val' separate
    print(f"{key} → {val}")

for item in data.items():             # loop variable is both 'key' and 'val', loop over entire dictionary, returns a tuple of key-values
    print(item)


# To iterate over the key-value pairs while keeping track of a counter, use enumerate() method
# enumerate() assigns an index or "count" to each key in the dictionary
for letter in enumerate(data):
    print(letter)

# You can assign values to separate loop variables i.e. 'key' and 'val'
for key, val in enumerate(data):
    print(key, val)

# Iterate over the values. Output will have index and the value in each tuple
for val in enumerate(data.values()):
    print(val)

# You can assign the 'val' to separate loop variables too
for index, val in enumerate(data.values()):
    print(index, val)

# With data.items() you can get the entire key-value pair along with an index/counter
for index, item in enumerate(data.items()):
    print(index, item)
```

## 11.4 Nested Dictionaries

Dictionaries can contain other dictionaries — great for representing complex structured data:

```python
students = {
    "alice": {"grade": "A", "score": 95, "passed": True},
    "bob":   {"grade": "B", "score": 82, "passed": True},
    "carol": {"grade": "F", "score": 45, "passed": False},
}

# Access nested values
print(students["alice"]["score"])     # 95

# Iterate
for name, data in students.items():
    status = "✓" if data["passed"] else "✗"
    print(f"{status} {name:8} — {data['grade']} ({data['score']})")
```
