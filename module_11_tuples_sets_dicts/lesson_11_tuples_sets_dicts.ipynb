# Module 11: Tuples, Sets & Dictionaries

Python has four primary collection types. This module covers the remaining three after lists.

## 11.1 Tuples — Immutable Sequences

Tuples are like lists, but **immutable** (cannot be changed after creation). Use them for data that should not change.

```python
# Creating tuples
coords = (3.5, 7.2)          # typical
rgb    = (255, 128, 0)
singleton = (42,)             # single item needs trailing comma!
nested = ((1, 2), (3, 4))

# Access (same as list)
print(coords[0])   # 3.5
print(rgb[-1])     # 0
print(len(rgb))    # 3

# Immutable — this would fail:
# coords[0] = 99  # TypeError!

# Packing and unpacking
point = 10, 20        # packing (parens optional)
x, y = point          # unpacking
print(x, y)           # 10 20

a, b, *rest = (1, 2, 3, 4, 5)
print(a, b, rest)     # 1 2 [3, 4, 5]
```

### When to use tuples vs lists

| Tuples | Lists |
|--------|-------|
| Fixed data (coordinates, RGB, DB rows) | Data that changes |
| Function return values | Working with a collection |
| Dictionary keys (they're hashable) | Not usable as dict keys |
| Slightly faster than lists | Slightly slower |

## 11.2 Sets — Unique Unordered Collections

A set is an **unordered collection of unique** values. Duplicate items are automatically removed.

```python
# Creating sets
colors = {"red", "green", "blue"}
nums   = {1, 2, 3, 4, 5}
empty  = set()           # NOT {} — that creates a dict!

# Duplicates are removed
dupes = {1, 2, 2, 3, 3, 3}
print(dupes)    # {1, 2, 3}

# Membership testing — very fast!
print("red" in colors)    # True
print("purple" in colors) # False

# Basic operations
colors.add("purple")
colors.remove("red")       # raises KeyError if not found
colors.discard("yellow")   # no error if not found
print(colors)
```

```python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7}

print("Union:       ", a | b)   # all elements from both
print("Intersection:", a & b)   # only elements in BOTH
print("Difference:  ", a - b)   # in a but NOT in b
print("Sym diff:    ", a ^ b)   # in one but NOT both

# Subset / superset
print({1,2}.issubset({1,2,3}))    # True
print({1,2,3}.issuperset({1,2}))  # True
```

## 11.3 Dictionaries — Key-Value Pairs

Dictionaries map **keys** to **values**. Keys must be unique and immutable (strings, numbers, tuples). Values can be anything.

```python
# Creating a dict
person = {
    "name":    "Alice",
    "age":     30,
    "city":    "London",
    "hobbies": ["reading", "coding"]
}

# Access by key
print(person["name"])       # Alice
print(person["hobbies"])    # ['reading', 'coding']

# .get() — safe access (returns None or default if key missing)
print(person.get("age"))          # 30
print(person.get("salary"))       # None
print(person.get("salary", 0))    # 0 (default value)

# Add or modify
person["email"] = "alice@example.com"
person["age"] = 31
print(person)
```

```python
# Remove entries
person = {"a": 1, "b": 2, "c": 3, "d": 4}

val  = person.pop("b")       # remove & return value
print(val, person)

del person["c"]              # delete entry
print(person)

last = person.popitem()      # remove & return last (k,v) pair
print(last, person)

# Iterate over dict
data = {"x": 10, "y": 20, "z": 30}

for key in data:                      # keys (default)
    print(key)

for val in data.values():             # values
    print(val)

for key, val in data.items():         # key-value pairs
    print(f"{key} → {val}")
```

```python
# Useful dict methods
d = {"a": 1, "b": 2}

print(d.keys())       # dict_keys(['a', 'b'])
print(d.values())     # dict_values([1, 2])
print(d.items())      # dict_items([('a', 1), ('b', 2)])

print(len(d))         # 2
print("a" in d)       # True  ← checks keys
print(1 in d.values()) # True ← checks values

# Merge dicts (Python 3.9+)
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}
merged = d1 | d2
print(merged)

# Update (add/overwrite)
d1.update({"b": 99, "e": 5})
print(d1)
```

## 11.4 Nested Dictionaries

Dictionaries can contain other dictionaries — great for representing complex structured data:

```python
students = {
    "alice": {"grade": "A", "score": 95, "passed": True},
    "bob":   {"grade": "B", "score": 82, "passed": True},
    "carol": {"grade": "F", "score": 45, "passed": False},
}

# Access nested values
print(students["alice"]["score"])     # 95

# Iterate
for name, data in students.items():
    status = "✓" if data["passed"] else "✗"
    print(f"{status} {name:8} — {data['grade']} ({data['score']})")
```

## ✅ Module 11 Summary
