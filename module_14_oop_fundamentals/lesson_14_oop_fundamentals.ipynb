# Module 14: Object-Oriented Programming — Fundamentals

OOP is a programming paradigm that organizes code around **objects** — combinations of data and behavior. Python is a fully object-oriented language.

## 14.1 Why OOP?

Procedural code (what we've written so far) works well for small programs. But as programs grow, OOP provides:

- **Encapsulation** — bundle related data and functions together
- **Reusability** — create many objects from one blueprint
- **Maintainability** — changes in one place affect all instances
- **Modeling** — represent real-world entities naturally

**Analogy:** A **class** is like a blueprint for a house. An **object** (instance) is an actual house built from that blueprint.

## 14.2 Defining a Class

````python
class Dog:
    """Represents a dog."""
    
    # Class attribute — shared by ALL instances
    species = "Canis familiaris"
    
    # __init__ is the constructor — called when creating an instance
    def __init__(self, name, breed, age):
        # Instance attributes — unique to each dog
        self.name  = name
        self.breed = breed
        self.age   = age
    
    # Instance method — always takes self as first param
    def bark(self):
        return f"{self.name} says: Woof!"
    
    def description(self):
        return f"{self.name} is a {self.age}-year-old {self.breed}"

# Creating instances
fido   = Dog("Fido",   "Labrador", 3)
bella  = Dog("Bella",  "Poodle",   5)
rex    = Dog("Rex",    "German Shepherd", 2)

print(fido.bark())
print(bella.description())
print(rex.name, rex.breed)
print(Dog.species)   # class attribute via class name
print(fido.species)  # class attribute via instance
````

## 14.3 The `self` Parameter

`self` is a reference to the **current instance**. It's automatically passed when you call an instance method — you don't pass it explicitly:

```python
fido.bark()   # Python translates to: Dog.bark(fido)
```

You can name it anything, but `self` is the universal convention.

## 14.4 Instance Attributes vs Class Attributes

````python
class Counter:
    # Class attribute — shared by all instances
    total_count = 0
    
    def __init__(self, name):
        self.name = name   # instance attribute
        Counter.total_count += 1
    
    def greet(self):
        print(f"Hello from counter #{Counter.total_count}: {self.name}")

c1 = Counter("Alpha")
c2 = Counter("Beta")
c3 = Counter("Gamma")

c1.greet()
c3.greet()
print(f"Total Counter objects created: {Counter.total_count}")
````

## 14.5 Methods — Types Overview

| Type | Decorator | First Param | Access |
|------|-----------|-------------|--------|
| Instance method | (none) | `self` | Instance data |
| Class method | `@classmethod` | `cls` | Class data |
| Static method | `@staticmethod` | (none) | Neither |

````python
class Temperature:
    """Temperature with unit conversion utilities."""
    
    def __init__(self, celsius):
        self.celsius = celsius
    
    def to_fahrenheit(self):        # instance method
        return self.celsius * 9/5 + 32
    
    @classmethod
    def from_fahrenheit(cls, f):    # class method — alternative constructor
        return cls((f - 32) * 5/9)
    
    @staticmethod
    def is_freezing(celsius):       # static method — utility, no self/cls
        return celsius <= 0
    
    def __repr__(self):
        return f"Temperature({self.celsius:.1f}°C)"

t1 = Temperature(100)
print(t1.to_fahrenheit())           # 212.0

t2 = Temperature.from_fahrenheit(32)
print(t2)                           # Temperature(0.0°C)

print(Temperature.is_freezing(-5))  # True
print(Temperature.is_freezing(20))  # False
````

## 14.6 Properties — Controlled Attribute Access

The `@property` decorator makes a method behave like an attribute, with optional setter/deleter:

````python
class BankAccount:
    def __init__(self, owner, initial_balance=0):
        self.owner   = owner
        self._balance = initial_balance   # _balance: "private by convention"
    
    @property
    def balance(self):                   # getter
        return self._balance
    
    @balance.setter
    def balance(self, amount):           # setter
        if amount < 0:
            raise ValueError("Balance cannot be negative")
        self._balance = amount
    
    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit must be positive")
        self._balance += amount
        return self._balance
    
    def __repr__(self):
        return f"BankAccount({self.owner!r}, ${self._balance:.2f})"

acct = BankAccount("Alice", 1000)
print(acct.balance)       # uses the getter
acct.deposit(500)
print(acct)               # uses __repr__
acct.balance = 200        # uses the setter
print(acct.balance)

try:
    acct.balance = -100   # ValueError from setter
except ValueError as e:
    print(f"Error: {e}")
````

## ✅ Module 14 Summary

| Concept | Description |
|---------|-------------|
| Class | Blueprint defined with `class Name:` |
| Instance | Object created from class: `obj = MyClass()` |
| `__init__` | Constructor — sets up instance attributes |
| `self` | Reference to current instance |
| Instance attribute | `self.x = value` — per-object |
| Class attribute | `ClassName.x = value` — shared |
| `@classmethod` | Takes `cls`, alternative constructors |
| `@staticmethod` | No `self`/`cls`, utility functions |
| `@property` | Controlled attribute access |
