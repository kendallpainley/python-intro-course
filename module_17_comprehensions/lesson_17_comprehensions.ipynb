# Module 17: Comprehensions

Comprehensions are concise, readable one-liners for creating collections. They are one of Python's most loved features.

## 17.1 List Comprehensions

```python
[expression for item in iterable]
[expression for item in iterable if condition]
```

They replace many common `for` loops that build a list.

````python
# Traditional for loop:
squares = []
for n in range(1, 11):
    squares.append(n ** 2)
print(squares)

# List comprehension — same result, one line:
squares = [n ** 2 for n in range(1, 11)]
print(squares)

# With a filter condition:
evens = [n for n in range(20) if n % 2 == 0]
print(evens)

# More complex expression:
words = ["hello", "world", "python", "is", "great"]
long_upper = [w.upper() for w in words if len(w) > 4]
print(long_upper)    # ['HELLO', 'WORLD', 'PYTHON', 'GREAT']
````

````python
# Practical examples

# Extract numbers from a mixed list
data = [1, "two", 3, "four", 5, "six"]
nums = [x for x in data if isinstance(x, int)]
print(nums)   # [1, 3, 5]

# Flatten a 2D list
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flat = [n for row in matrix for n in row]
print(flat)   # [1,2,3,4,5,6,7,8,9]

# Cartesian product
colors = ["red", "blue"]
sizes  = ["S", "M", "L"]
products = [(c, s) for c in colors for s in sizes]
print(products)

# Conditional expression (ternary) in comprehension
labels = ["even" if n % 2 == 0 else "odd" for n in range(8)]
print(labels)
````

## 17.2 Dictionary Comprehensions

```python
{key_expr: value_expr for item in iterable}
```

````python
# Create a dict of squares
squares_dict = {n: n**2 for n in range(1, 8)}
print(squares_dict)   # {1: 1, 2: 4, 3: 9, ...}

# Invert a dictionary (swap keys and values)
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(inverted)       # {1: "a", 2: "b", 3: "c"}

# Filter a dictionary
scores = {"Alice": 92, "Bob": 74, "Carol": 88, "Dave": 65}
high_scorers = {name: s for name, s in scores.items() if s >= 80}
print(high_scorers)   # Alice and Carol only

# Transform values
normalized = {k: v/100 for k, v in scores.items()}
print(normalized)
````

## 17.3 Set Comprehensions

```python
{expression for item in iterable}
```

````python
# Set of unique lengths of words
words = ["apple", "banana", "cherry", "date", "elderberry", "fig"]
unique_lengths = {len(w) for w in words}
print(unique_lengths)    # {3, 4, 5, 6, 10}  (order varies)

# Unique first letters
first_letters = {w[0] for w in words}
print(first_letters)

# Filter + transform
even_squares = {n**2 for n in range(20) if n % 2 == 0}
print(sorted(even_squares))
````

## 17.4 Generator Expressions

Like list comprehensions but with `()` instead of `[]`. They produce values **lazily** (one at a time) instead of building a whole list in memory.

Use generators when:
- The sequence is very large
- You only need to iterate once
- You're passing to a function like `sum()`, `max()`, `any()`

````python
# List comprehension — builds entire list in memory:
squares_list = [n**2 for n in range(1_000_000)]

# Generator expression — produces values one at a time:
squares_gen = (n**2 for n in range(1_000_000))

print(type(squares_list))   # <class 'list'>
print(type(squares_gen))    # <class 'generator'>

# Generators work great with sum(), max(), any(), all()
total = sum(n**2 for n in range(1001))   # no [] needed
print(f"Sum of squares 0-1000: {total}")

has_negative = any(n < 0 for n in [1, 2, -3, 4])
print(f"Has negative: {has_negative}")   # True

all_positive = all(n > 0 for n in [1, 2, 3, 4])
print(f"All positive: {all_positive}")   # True
````

## 17.5 When NOT to Use Comprehensions

Comprehensions are great, but can be overused. Avoid them when:

- The logic is complex (more than 2 conditions/levels)
- Side effects are needed inside the loop
- Debugging is difficult

```python
# Too complex — use a regular loop instead:
# bad = [process(x) for x in items if predicate(x) for y in x if other(y)]
```

````python
# Example: comparing readability

# Nested comprehension (hard to read):
matrix = [[1,2,3],[4,5,6],[7,8,9]]
transposed = [[row[i] for row in matrix] for i in range(3)]
print("Transposed:", transposed)

# Same thing as a loop (clearer for beginners):
transposed2 = []
for i in range(3):
    col = []
    for row in matrix:
        col.append(row[i])
    transposed2.append(col)
print("Same:      ", transposed2)
````

| Dict | `{k: v for x in it}` | `dict` |
| Set | `{expr for x in it}` | `set` |
| Generator | `(expr for x in it)` | `generator` |
