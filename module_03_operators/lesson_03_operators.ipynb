# Module 03: Operators

Operators let you perform computations, make comparisons, and combine logic. Python has five main categories.

## 3.1 Arithmetic Operators

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | Division (float) | `7 / 2` | `3.5` |
| `//` | Floor division | `7 // 2` | `3` |
| `%` | Modulo (remainder) | `7 % 2` | `1` |
| `**` | Exponentiation | `2 ** 8` | `256` |

````python
print(10 + 3)   # 13
print(10 - 3)   # 7
print(10 * 3)   # 30
print(10 / 3)   # 3.3333... (always float)
print(10 // 3)  # 3  (floor — rounds DOWN)
print(10 % 3)   # 1  (remainder)
print(2 ** 10)  # 1024

# Practical examples
minutes = 137
hours = minutes // 60
remaining = minutes % 60
print(f"{minutes} minutes = {hours}h {remaining}m")  # 2h 17m
````

## 3.2 Comparison Operators

Comparisons always return `True` or `False`:

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal | `5 != 3` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<` | Less than | `5 < 3` | `False` |
| `>=` | Greater or equal | `5 >= 5` | `True` |
| `<=` | Less or equal | `4 <= 3` | `False` |

> ⚠️ **Common mistake:** `=` is *assignment*, `==` is *comparison*!

````python
x = 10

print(x == 10)   # True
print(x == 5)    # False
print(x != 5)    # True
print(x > 7)     # True
print(x < 7)     # False
print(x >= 10)   # True
print(x <= 9)    # False

# Comparing strings
print("apple" == "apple")   # True
print("apple" == "Apple")   # False (case-sensitive!)
print("banana" > "apple")   # True (alphabetical order)
````

## 3.3 Logical Operators

Combine multiple conditions:

| Operator | Meaning | Result |
|----------|---------|--------|
| `and` | Both must be True | `True and True` → `True` |
| `or` | At least one True | `False or True` → `True` |
| `not` | Inverts the value | `not True` → `False` |

### Truth table for `and` and `or`
```
A       B       A and B   A or B
─────────────────────────────────
True    True    True      True
True    False   False     True
False   True    False     True
False   False   False     False
```

````python
age = 20
has_id = True
is_member = False

# and: both conditions must be True
print(age >= 18 and has_id)        # True
print(age >= 18 and is_member)     # False

# or: at least one must be True
print(age >= 18 or is_member)      # True
print(age < 18 or is_member)       # False

# not: flip the boolean
print(not is_member)               # True
print(not has_id)                  # False

# Combined
can_enter = age >= 18 and (has_id or is_member)
print("Can enter:", can_enter)     # True
````

## 3.4 Assignment Operators

Shorthand for updating a variable's value:

| Operator | Equivalent | Example |
|----------|------------|---------|
| `+=` | `x = x + n` | `x += 5` |
| `-=` | `x = x - n` | `x -= 3` |
| `*=` | `x = x * n` | `x *= 2` |
| `/=` | `x = x / n` | `x /= 4` |
| `//=` | `x = x // n` | `x //= 2` |
| `%=` | `x = x % n` | `x %= 7` |
| `**=` | `x = x ** n` | `x **= 3` |

````python
score = 100
print(score)    # 100

score += 50     # score = score + 50
print(score)    # 150

score -= 20
print(score)    # 130

score *= 2
print(score)    # 260

score //= 3
print(score)    # 86
````

## 3.5 Operator Precedence

When multiple operators appear, Python follows **PEMDAS** order (same as maths):

1. `()` Parentheses
2. `**` Exponentiation
3. `*`, `/`, `//`, `%` Multiplication/Division
4. `+`, `-` Addition/Subtraction
5. `==`, `!=`, `>`, `<`, `>=`, `<=` Comparisons
6. `not` Logical NOT
7. `and` Logical AND
8. `or` Logical OR

**Rule of thumb:** When in doubt, use parentheses to make intent explicit.

````python
# Without parentheses:
result = 2 + 3 * 4
print(result)    # 14 (not 20! — * before +)

# With parentheses to control order:
result = (2 + 3) * 4
print(result)    # 20

# Complex expression:
x = 5
print(x ** 2 + 2 * x - 1)         # 25 + 10 - 1 = 34
print(x ** 2 + 2 * (x - 1))       # 25 + 8 = 33
````

## 3.6 Identity and Membership Operators

| Operator | Meaning | Example |
|----------|---------|----------|
| `is` | Same object in memory | `x is None` |
| `is not` | Not the same object | `x is not None` |
| `in` | Value exists in collection | `3 in [1,2,3]` |
| `not in` | Value NOT in collection | `5 not in [1,2,3]` |

````python
# Identity operators — use "is" for None checks
value = None
print(value is None)      # True
print(value is not None)  # False

# Membership operators — check if item is in a sequence
fruits = ["apple", "banana", "cherry"]
print("banana" in fruits)      # True
print("mango" in fruits)       # False
print("mango" not in fruits)   # True

# Works on strings too:
sentence = "Hello, World!"
print("World" in sentence)     # True
print("Python" in sentence)    # False
````
