## 16.1 Importing Modules

A **module** is any `.py` file. A **package** is a directory of modules.

````python
# Import the whole module
import math
print(math.sqrt(16))


# Import specific names from a module
from math import sqrt
print(sqrt(25))

# Import specific names assigned to aliases
from math import pi as p
print(p)

# Import with an alias
import numpy as np
print(np.sqrt(36))

# Import entire module using *, but you don't have to use the prefix of the module i.e. math.sqrt() becomes sqrt() etc.
from math import *
print(exp(1))
````

## 16.2 Key Standard Library Modules

The standard library ships with Python

| Module | Use |
|--------|-----|
| `math` | Maths functions |
| `random` | Random numbers |
| `datetime` | Dates and times |
| `os` | OS interface, file operations |
| `sys` | Python runtime info |
| `json` | JSON encode/decode |
| `re` | Regular expressions |
| `collections` | Specialized containers |
| `itertools` | Iterator tools |
| `functools` | Functional utilities |

````python
import random

# Random integers
print(random.randint(1, 10))      # 1-10 inclusive
print(random.randrange(0, 100, 5)) # 0,5,10,...,95

# Random floats
print(random.random())             # 0.0 to 1.0
print(random.uniform(1.5, 3.0))   # 1.5 to 3.0

# Random from sequence
fruit = random.choice(["apple","banana","cherry"])
print(fruit)

# Shuffle a list (in-place)
deck = list(range(1, 11))
random.shuffle(deck)
print(deck)

# Sample without replacement
print(random.sample(range(100), 5))  # 5 unique numbers

# Reproducible randomness
random.seed(42)
print(random.random())   # always same value with seed 42
````

````python
import datetime

# Current date and time
now = datetime.datetime.now()
print(now)
print(f"Year: {now.year}, Month: {now.month}, Day: {now.day}")
print(f"Hour: {now.hour}, Minute: {now.minute}")

# Create specific dates
birthday = datetime.date(1990, 5, 15)
today = datetime.date.today()
delta = today - birthday
print(f"Days since birthday: {delta.days}")

# Formatting
print(now.strftime("%Y-%m-%d %H:%M:%S"))
print(now.strftime("%A, %B %d, %Y"))  # Tuesday, March 15, 2024

# Parsing a date string
date_str = "2024-12-25"
parsed = datetime.datetime.strptime(date_str, "%Y-%m-%d")
print(parsed)

# Time deltas
one_week = datetime.timedelta(weeks=1)
next_week = today + one_week
print(f"Next week: {next_week}")
````

## 16.3 `os` and `sys`

````python
import os, sys

# Current working directory
print(os.getcwd())

# Environment variables
path = os.getenv("PATH", "not set")
print(f"PATH: {path[:50]}...")

# Join paths safely
fullpath = os.path.join("/home", "user", "docs", "file.txt")
print(fullpath)

# Check existence
print(os.path.exists("/tmp"))     # True
print(os.path.isdir("/tmp"))      # True
print(os.path.isfile("/tmp"))     # False

# sys info
print(sys.version)
print(sys.platform)
print(f"Python executable: {sys.executable}")
````

## 16.4 `json` — Serialize Python Objects

````python
import json

# Python → JSON string
data = {
    "name": "Alice",
    "age": 30,
    "scores": [95, 87, 92],
    "active": True,
    "address": None
}

json_str = json.dumps(data, indent=2)
print(json_str)

# JSON string → Python
parsed = json.loads(json_str)
print(type(parsed))        # dict
print(parsed["name"])      # Alice

# Write to file
with open("data.json", "w") as f:
    json.dump(data, f, indent=2)

# Read from file
with open("data.json") as f:
    loaded = json.load(f)
print(loaded)
````

## 16.6 Virtual Environments and pip

A **virtual environment** is an isolated Python installation per project, it is essential for managing dependencies.

```bash
# Create a virtual environment
python -m venv venv

# Activate it
source venv/bin/activate      # Mac/Linux
venv\Scripts\activate         # Windows

# Install packages
pip install requests pandas numpy

# Save dependencies
pip freeze > requirements.txt

# Install from requirements.txt (on another machine)
pip install -r requirements.txt

# Deactivate
deactivate
```

In PyCharm: Settings → Project → Python Interpreter → Add Interpreter → Virtualenv Environment.
