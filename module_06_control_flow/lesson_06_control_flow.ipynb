# Module 06: Control Flow — if / elif / else

Control flow determines which code runs and when. Python's primary tool for decision-making is the `if` statement.

## 6.1 The `if` Statement

Syntax:
```python
if condition:
    # code that runs when condition is True
```

The colon `:` and indentation are **required**. The body must be indented (4 spaces by convention).

````python
temperature = 35

if temperature > 30:
    print("It's hot outside!")
    print("Remember to stay hydrated.")

print("This always runs, regardless of temperature.")
````

## 6.2 `if` / `else`

```python
if condition:
    # runs when True
else:
    # runs when False
```

````python
age = 16

if age >= 18:
    print("You may vote.")
else:
    print("You are not old enough to vote yet.")

# The else block always runs when the condition is False
balance = 0
if balance > 0:
    print(f"Balance: ${balance}")
else:
    print("Your account is empty.")
````

## 6.3 `if` / `elif` / `else`

For multiple conditions, use `elif` (short for "else if"). Python checks conditions top to bottom and runs the **first True block only**.

````python
score = 75

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")
````

````python
# Day of the week
day_num = 3  # 1 = Monday, 7 = Sunday

if day_num == 1:
    day = "Monday"
elif day_num == 2:
    day = "Tuesday"
elif day_num == 3:
    day = "Wednesday"
elif day_num == 4:
    day = "Thursday"
elif day_num == 5:
    day = "Friday"
elif day_num in (6, 7):
    day = "Weekend!"
else:
    day = "Invalid day number"

print(day)
````

## 6.4 Nested Conditions

You can place `if` statements inside other `if` blocks:

````python
age = 20
has_id = True
is_vip = False

if age >= 18:
    print("Age check passed.")
    if has_id:
        print("ID verified.")
        if is_vip:
            print("Welcome to the VIP lounge!")
        else:
            print("Welcome to the general area.")
    else:
        print("No ID — cannot enter.")
else:
    print("Underage — entry refused.")

# Tip: keep nesting shallow (max 2-3 levels) for readability.
# Often you can replace nesting with `and` / `or`.
````

## 6.5 Ternary Expression (Conditional Expression)

A compact one-line `if/else` for simple assignments:

```python
value = <true_result> if <condition> else <false_result>
```

````python
# Traditional if/else:
age = 20
if age >= 18:
    status = "adult"
else:
    status = "minor"
print(status)

# Ternary version (same logic, one line):
status = "adult" if age >= 18 else "minor"
print(status)

# Practical uses:
num = -5
abs_val = num if num >= 0 else -num
print(f"|{num}| = {abs_val}")

temp = 22
clothing = "shorts" if temp >= 20 else "jacket"
print(f"Wear {clothing}!")
````

## 6.6 `match` Statement (Python 3.10+)

The `match` statement is Python's pattern matching — cleaner than long `elif` chains for comparing a value against multiple options:

````python
command = "quit"

match command:
    case "start":
        print("Starting the program...")
    case "stop" | "quit" | "exit":  # ← multiple options with |
        print("Shutting down.")
    case "help":
        print("Available commands: start, stop, quit, help")
    case _:                          # ← default (like else)
        print(f"Unknown command: {command}")
````

## 6.7 Common Patterns

````python
# 1. Validation pattern
username = "alice"
if not username:
    print("Username cannot be empty!")
elif len(username) < 3:
    print("Username too short (min 3 chars).")
elif len(username) > 20:
    print("Username too long (max 20 chars).")
else:
    print(f"Username '{username}' is valid!")

# 2. Guard clause (early return pattern)
def check_age(age):
    if age < 0:
        print("Age cannot be negative!")
        return
    if age > 150:
        print("Age seems unrealistic!")
        return
    print(f"Age {age} is valid.")

check_age(25)
check_age(-5)
check_age(200)
````
