# Module 05: User Input & Type Conversion

Programs become truly interactive when they can read input from users and convert between types.

## 5.1 The `input()` Function

`input()` pauses the program, displays a prompt, and waits for the user to type something and press Enter.

**Important:** `input()` **always returns a string**, even if the user types a number.

```python
# Basic usage
name = input("What is your name? ")
print(f"Hello, {name}!")

# The prompt is optional
favourite = input("Your favourite colour: ")
print(f"{favourite} is a great choice!")

# input() always returns str
age_str = input("Your age: ")
print(type(age_str))   # <class 'str'>
```

## 5.2 Type Conversion (Casting)

Since `input()` returns a string, you often need to convert it to the right type.

| Function | Converts to | Example |
|----------|-------------|----------|
| `int(x)` | Integer | `int("42")` → `42` |
| `float(x)` | Float | `float("3.14")` → `3.14` |
| `str(x)` | String | `str(99)` → `"99"` |
| `bool(x)` | Boolean | `bool(0)` → `False` |
| `list(x)` | List | `list("abc")` → `['a','b','c']` |

```python
# Convert input to a number before doing maths
age = int(input("Enter your age: "))
next_birthday = age + 1
print(f"Next year you will be {next_birthday}!")

# Convert to float for decimal input
price = float(input("Price: $"))
print(f"With 10% tax: ${price * 1.10:.2f}")

# Convert numbers to strings for string operations
year = 2024
print("Year: " + str(year))  # must convert to str first
```

## 5.3 Truthy and Falsy Values

`bool()` converts any value to `True` or `False`. Python considers these values **falsy** (behave as `False`):

- `0`, `0.0` (zero)
- `""` (empty string)
- `[]`, `{}`, `()`, `set()` (empty collections)
- `None`

Everything else is **truthy** (behaves as `True`).

```python
# Falsy values:
print(bool(0))      # False
print(bool(""))     # False
print(bool(None))   # False
print(bool([]))     # False

# Truthy values:
print(bool(1))          # True
print(bool(-5))         # True  (non-zero)
print(bool("hello"))    # True  (non-empty)
print(bool([1, 2]))     # True  (non-empty)
print(bool(0.001))      # True
```

## 5.4 Handling Conversion Errors

If you try to convert an invalid string, Python raises a `ValueError`. We'll cover error handling fully in Module 13, but here's a preview:

```python
# This would crash if user types "abc":
# age = int(input("Age: "))  # ValueError if non-numeric

# Safe approach using try/except (preview of Module 13):
try:
    value = int("hello")  # simulate bad input
except ValueError:
    print("That's not a valid integer!")

try:
    value = int("42")
    print(f"Successfully converted: {value}")
except ValueError:
    print("Conversion failed!")
```

## 5.5 Type Conversion Between Numbers

```python
# int to float
x = int(3.9)     # truncates (does NOT round!)
print(x)         # 3

y = int(-3.9)
print(y)         # -3  (truncates toward zero)

# float to int
z = float(7)
print(z)         # 7.0

# round() for actual rounding
print(round(3.7))     # 4
print(round(3.2))     # 3
print(round(3.456, 2))  # 3.46 (2 decimal places)
```

## 5.6 Practical Example — Simple Calculator

```python
print("=== Simple Calculator ===")
num1 = float(input("First number: "))
operator = input("Operator (+, -, *, /): ")
num2 = float(input("Second number: "))

if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/" and num2 != 0:
    result = num1 / num2
else:
    result = None
    print("Invalid operator or division by zero!")

if result is not None:
    print(f"{num1} {operator} {num2} = {result}")
```

## ✅ Module 05 Summary

| Concept | Key point |
|---------|----------|
| `input()` | Always returns a `str` |
| `int()` | Converts to integer (truncates floats) |
| `float()` | Converts to decimal |
| `str()` | Converts anything to string |
| `bool()` | 0, `""`, `None`, empty = `False`; all else = `True` |
| `ValueError` | Raised when conversion is impossible |
