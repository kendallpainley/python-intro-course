# Module 13: Error Handling & Exceptions

Errors are inevitable. Python's exception system lets you handle them gracefully instead of crashing.

## 13.1 Types of Errors

| Error Type | When it occurs | Example |
|-----------|--------------|----------|
| `SyntaxError` | Code doesn't parse | `if x` (missing colon) |
| `NameError` | Variable doesn't exist | `print(x)` (x not defined) |
| `TypeError` | Wrong type | `1 + "a"` |
| `ValueError` | Right type, bad value | `int("hello")` |
| `IndexError` | List index out of range | `lst[99]` |
| `KeyError` | Dict key missing | `d["x"]` (x not in d) |
| `ZeroDivisionError` | Division by zero | `1 / 0` |
| `FileNotFoundError` | File missing | `open("nope.txt")` |
| `AttributeError` | Object has no attribute | `None.upper()` |

````python
# Each of these would raise an exception:
try:
    print(undefined_var)      # NameError
except NameError as e:
    print(f"NameError: {e}")

try:
    x = 1 + "two"            # TypeError
except TypeError as e:
    print(f"TypeError: {e}")

try:
    x = int("hello")         # ValueError
except ValueError as e:
    print(f"ValueError: {e}")

try:
    x = 10 / 0               # ZeroDivisionError
except ZeroDivisionError as e:
    print(f"ZeroDivisionError: {e}")
````

## 13.2 `try` / `except`

```python
try:
    # code that might fail
except ExceptionType:
    # handle the error
```

````python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return None

print(safe_divide(10, 2))   # 5.0
print(safe_divide(10, 0))   # None

# Handle multiple exceptions
def parse_number(text):
    try:
        return int(text)
    except ValueError:
        try:
            return float(text)
        except ValueError:
            return None

print(parse_number("42"))      # 42
print(parse_number("3.14"))    # 3.14
print(parse_number("hello"))   # None
````

## 13.3 `except` with Multiple Types

````python
def read_value(data, key):
    try:
        value = data[key]
        return int(value)
    except KeyError:
        print(f"Key {key!r} not found")
    except (TypeError, ValueError) as e:
        print(f"Conversion error: {e}")
    return None

print(read_value({"x": "42"}, "x"))      # 42
print(read_value({"x": "hi"}, "x"))      # Conversion error
print(read_value({"x": "42"}, "y"))      # Key not found
````

## 13.4 `else` and `finally`

```python
try:
    # risky code
except SomeError:
    # handle error
else:
    # runs ONLY if no exception occurred
finally:
    # ALWAYS runs — cleanup here
```

````python
def read_file(filename):
    f = None
    try:
        f = open(filename)
        content = f.read()
    except FileNotFoundError:
        print(f"File not found: {filename}")
        content = None
    else:
        print(f"Successfully read {len(content)} characters")
    finally:
        if f:
            f.close()   # always close the file
        print("Cleanup done")
    return content

result = read_file("sample.txt")   # exists from module 12
result = read_file("missing.txt")
````

## 13.5 `raise` — Triggering Exceptions

````python
def set_age(age):
    if not isinstance(age, int):
        raise TypeError(f"Age must be an int, got {type(age).__name__}")
    if age < 0 or age > 150:
        raise ValueError(f"Age {age} is out of valid range (0-150)")
    return age

try:
    set_age(25)
    print("Age 25 accepted")
except (TypeError, ValueError) as e:
    print(f"Error: {e}")

try:
    set_age(-5)
except ValueError as e:
    print(f"Error: {e}")

# re-raise after logging
try:
    set_age("old")
except TypeError as e:
    print(f"Logging: {e}")
    raise   # re-raises the same exception
````

## 13.6 Custom Exceptions

Create your own exception classes by subclassing `Exception`:

````python
class InsufficientFundsError(Exception):
    """Raised when a bank account has insufficient funds."""
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(
            f"Cannot withdraw ${amount:.2f}: balance is ${balance:.2f}"
        )

class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientFundsError(self.balance, amount)
        self.balance -= amount
        return self.balance

account = BankAccount(100)
print(account.withdraw(30))    # 70

try:
    account.withdraw(200)
except InsufficientFundsError as e:
    print(f"Error: {e}")
````

## ✅ Module 13 Summary

| Keyword | Purpose |
|---------|---------|
| `try:` | Wrap risky code |
| `except ExcType:` | Handle specific exception |
| `except (A,B):` | Handle multiple types |
| `except ExcType as e:` | Capture the error object |
| `else:` | Runs only if no exception |
| `finally:` | Always runs (cleanup) |
| `raise` | Trigger an exception |
| Custom exceptions | Subclass `Exception` |
