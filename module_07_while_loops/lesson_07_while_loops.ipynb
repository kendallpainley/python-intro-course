# Module 07: While Loops

Loops let you repeat code. The `while` loop repeats as long as a condition is `True`.

## 7.1 Basic `while` Loop

```python
while condition:
    # body — repeats while condition is True
```

The loop checks the condition **before** each iteration. When the condition becomes `False`, the loop stops.

````python
count = 1

while count <= 5:
    print(f"Count: {count}")
    count += 1  # IMPORTANT: update the variable or loop runs forever!

print("Done!")
````

## 7.2 Infinite Loops and `break`

A loop without a way to stop runs forever — called an **infinite loop**. Use `break` to exit a loop immediately from inside the body:

````python
# Infinite loop with break — common pattern for menus
while True:
    command = input("Enter command ('quit' to stop): ")
    if command == "quit":
        print("Exiting...")
        break
    print(f"You entered: {command}")
print("Loop ended.")
````

## 7.3 `continue` — Skip to Next Iteration

`continue` skips the rest of the current iteration and jumps back to the condition check:

````python
# Print only odd numbers from 1-10
num = 0
while num < 10:
    num += 1
    if num % 2 == 0:  # if even...
        continue      # ...skip the print
    print(num)
````

## 7.4 `while` / `else`

Python allows an `else` clause on loops. It runs when the loop finishes **normally** (condition became False), but NOT if the loop was exited with `break`:

````python
# Search for a value
target = 7
n = 1

while n <= 10:
    if n == target:
        print(f"Found {target}!")
        break
    n += 1
else:
    print(f"{target} not found in range 1-10.")

# Now search for something that isn't there:
target = 15
n = 1
while n <= 10:
    if n == target:
        print(f"Found {target}!")
        break
    n += 1
else:
    print(f"{target} not found in range 1-10.")  # This runs
````

## 7.5 Input Validation Pattern

One of the most useful `while` patterns: keep asking until you get valid input:

````python
# Validate a number in range
while True:
    try:
        age = int(input("Enter your age (1-120): "))
        if 1 <= age <= 120:
            break  # valid — exit the loop
        else:
            print("Please enter a number between 1 and 120.")
    except ValueError:
        print("That's not a valid number. Try again.")

print(f"Your age is: {age}")
````

## 7.6 Countdown Timer

````python
import time

def countdown(seconds):
    while seconds > 0:
        print(f"  {seconds}...")
        # time.sleep(1)  # uncomment for a real 1-second delay
        seconds -= 1
    print("  BLAST OFF! 🚀")

countdown(5)
````

## 7.7 Accumulator Pattern

A common pattern: start with a base value and update it inside a loop:

````python
# Sum of numbers 1 to 100
total = 0
n = 1
while n <= 100:
    total += n
    n += 1
print(f"Sum 1-100 = {total}")  # 5050

# Factorial: n! = n * (n-1) * ... * 1
def factorial(n):
    result = 1
    while n > 1:
        result *= n
        n -= 1
    return result

print(f"5! = {factorial(5)}")   # 120
print(f"10! = {factorial(10)}") # 3628800
````

## ✅ Module 07 Summary

| Concept | Key point |
|---------|----------|
| `while condition:` | Repeats while condition is True |
| `break` | Exit loop immediately |
| `continue` | Skip to next iteration |
| `while/else` | `else` runs if loop didn't `break` |
| Infinite loop | `while True:` — needs explicit `break` |
