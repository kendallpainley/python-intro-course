# Module 12: File I/O

Python makes reading and writing files straightforward. This module covers text files, CSV files, and working with file paths.

## 12.1 Opening and Reading Files

Use `open()` with a **mode** flag:

| Mode | Meaning |
|------|---------|
| `'r'` | Read (default) |
| `'w'` | Write (creates or overwrites) |
| `'a'` | Append (add to end) |
| `'x'` | Create (fail if exists) |
| `'rb'` / `'wb'` | Binary read/write |

```python
# The WITH statement — always use it for files!
# It automatically closes the file even if an error occurs.

# Writing a file first so we have something to read:
with open("sample.txt", "w") as f:
    f.write("Hello, World!\n")
    f.write("Python file I/O is easy.\n")
    f.write("Third line here.\n")

print("File written!")

# Reading the entire file as one string
with open("sample.txt", "r") as f:
    content = f.read()
print("--- read() ---")
print(content)
```

```python
# readline() — one line at a time
with open("sample.txt") as f:  # mode defaults to "r"
    line1 = f.readline()
    line2 = f.readline()
print(repr(line1))    # note the \n at the end
print(repr(line2))

# readlines() — all lines as a list
with open("sample.txt") as f:
    lines = f.readlines()
print(lines)

# Iterating line by line (most memory-efficient)
print("--- Line by line ---")
with open("sample.txt") as f:
    for line in f:
        print(line.strip())    # strip() removes the \n
```

## 12.2 Writing Files

```python
# Write mode ("w") — creates or OVERWRITES
with open("output.txt", "w") as f:
    f.write("Line 1\n")
    f.write("Line 2\n")
    f.writelines(["Line 3\n", "Line 4\n", "Line 5\n"])

print("Written!")

# Append mode ("a") — adds to existing file
with open("output.txt", "a") as f:
    f.write("Line 6 (appended)\n")

# Verify
with open("output.txt") as f:
    print(f.read())
```

## 12.3 Working with File Paths — `pathlib`

`pathlib.Path` is the modern, OS-agnostic way to work with file paths:

```python
from pathlib import Path

# Create a path object
p = Path("output.txt")

print(p.name)       # output.txt
print(p.stem)       # output (name without extension)
print(p.suffix)     # .txt
print(p.parent)     # .  (current directory)
print(p.exists())   # True if file exists
print(p.is_file())  # True
print(p.is_dir())   # False

# Read/write directly from Path objects
p.write_text("Hello from pathlib!\n")
content = p.read_text()
print(content)

# Build paths in an OS-safe way
home = Path.home()
docs = home / "Documents"   # / operator joins paths
print(docs)
```

```python
from pathlib import Path

# Create a directory
Path("test_dir").mkdir(exist_ok=True)

# Write multiple files
for i in range(3):
    filepath = Path("test_dir") / f"file_{i}.txt"
    filepath.write_text(f"Content of file {i}\n")

# List files in a directory
for f in Path("test_dir").iterdir():
    print(f.name)

# Glob — find files by pattern
for f in Path("test_dir").glob("*.txt"):
    print(f"  {f.name}: {f.stat().st_size} bytes")

# Cleanup
import shutil
shutil.rmtree("test_dir")
```

## 12.4 CSV Files

```python
import csv

# Write a CSV file
students = [
    ["Name", "Age", "Grade"],
    ["Alice", 20, "A"],
    ["Bob", 22, "B"],
    ["Carol", 21, "A"],
]

with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(students)

print("CSV written!")

# Read a CSV file
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)

print()

# Use DictReader for named columns
with open("students.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"{row['Name']} is {row['Age']} years old, grade {row['Grade']}.")
```

## ✅ Module 12 Summary

| `for line in f:` | Most efficient iteration |
