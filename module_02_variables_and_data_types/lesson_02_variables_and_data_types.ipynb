# Module 02: Variables & Data Types

In this module you'll learn how Python stores information in **variables** and the fundamental **data types** that exist in Python.

## 2.1 What Is a Variable?

A **variable** is a named container that stores a value in memory.

Think of a variable like a labelled box:
```
┌─────────────┐
│    age      │  ← the label (variable name)
│─────────────│
│     25      │  ← the value stored inside
└─────────────┘
```
When you need the value, you refer to it by name.

````python
# Creating a variable: name = value
age = 25
name = "Alice"
height = 5.7
is_student = True

# Using variables
print(name)
print(age)
print(height)
print(is_student)
````

## 2.2 Variable Naming Rules

Python variable names must follow these rules:

| Rule | Valid | Invalid |
|------|-------|---------|
| Letters, digits, underscores only | `my_var`, `x1` | `my-var`, `x!` |
| Cannot start with a digit | `name2` | `2name` |
| Case-sensitive | `age` ≠ `Age` | — |
| Cannot be a keyword | `total` | `for`, `if`, `class` |

### Naming convention: `snake_case`
Python style (PEP 8) uses **snake_case** — all lowercase with underscores:

````python
# ✓ Good variable names (snake_case)
first_name = "Bob"
total_score = 100
is_logged_in = False
max_retry_count = 3

# ✗ Avoid these styles (they work but break convention)
firstName = "Bob"    # camelCase — used in JS, not Python
FirstName = "Bob"    # PascalCase — reserved for class names
TOTAL = 100          # ALL_CAPS — reserved for constants

print(first_name, total_score, is_logged_in)
````

## 2.3 Core Data Types

Python has several built-in data types. The four most fundamental are:

| Type | Keyword | Example | Description |
|------|---------|---------|-------------|
| Integer | `int` | `42` | Whole numbers |
| Float | `float` | `3.14` | Decimal numbers |
| String | `str` | `"hello"` | Text |
| Boolean | `bool` | `True` / `False` | True or False |

### Integers (`int`)

Whole numbers — positive, negative, or zero. No decimal point.

````python
x = 42
y = -100
z = 0
big = 1_000_000   # underscores improve readability (same as 1000000)

print(x, y, z, big)
print(type(x))    # <class 'int'>
````

### Floats (`float`)

Numbers with a decimal point. Used for measurements, percentages, etc.

````python
pi = 3.14159
temperature = -7.5
percentage = 0.75

print(pi, temperature, percentage)
print(type(pi))   # <class 'float'>

# Scientific notation also works
small = 1.5e-4    # 0.00015
large = 2.0e6     # 2000000.0
print(small, large)
````

### Strings (`str`)

Text data. Can use single quotes, double quotes, or triple quotes:

````python
greeting = "Hello, World!"
name = 'Alice'              # single quotes work too
message = "It's a great day!"  # double quotes allow apostrophes

# Triple quotes for multi-line strings
poem = """
Roses are red,
Violets are blue,
Python is great,
And so are you!
"""

print(greeting)
print(name)
print(message)
print(poem)
````

### Booleans (`bool`)

Exactly two values: `True` or `False`. Note the capital first letter — this is required.

````python
is_raining = True
has_ticket = False
is_adult = True

print(is_raining)
print(type(is_raining))  # <class 'bool'>

# Booleans are the result of comparisons:
print(5 > 3)   # True
print(10 == 7) # False
print(4 != 4)  # False
````

## 2.4 The `None` Type

`None` represents the **absence of a value** — like `null` in other languages.

It's useful for:
- Initializing a variable before you know its value
- Return value of functions that don't explicitly return anything

````python
result = None
user_input = None

print(result)           # None
print(type(result))     # <class 'NoneType'>
print(result is None)   # True  ← correct way to check for None
````

## 2.5 The `type()` Function

Use `type()` to find out what type a variable holds:

````python
print(type(42))        # <class 'int'>
print(type(3.14))      # <class 'float'>
print(type("hello"))   # <class 'str'>
print(type(True))      # <class 'bool'>
print(type(None))      # <class 'NoneType'>

# Or check a variable:
x = "Python"
print(type(x))         # <class 'str'>
````

## 2.6 Dynamic Typing

Python is **dynamically typed** — a variable can hold any type, and you can reassign it to a different type at any time.

This differs from statically typed languages like Java or C++, where a variable's type is fixed:

````python
x = 10
print(x, type(x))   # 10 <class 'int'>

x = "hello"
print(x, type(x))   # hello <class 'str'>

x = True
print(x, type(x))   # True <class 'bool'>

# While flexible, be careful: reassigning types unexpectedly
# can make code confusing. Use clear names and be consistent.
````

## 2.7 Multiple Assignment

Python supports several shorthand ways to assign variables:

````python
# Assign the same value to multiple variables
a = b = c = 0
print(a, b, c)    # 0 0 0

# Assign multiple variables in one line (tuple unpacking)
x, y, z = 1, 2, 3
print(x, y, z)    # 1 2 3

# Swap variables without a temporary variable!
a, b = 10, 20
print(a, b)       # 10 20
a, b = b, a
print(a, b)       # 20 10  ← swapped!
````

## 2.8 Constants

Python has no built-in constant type, but the convention is to name constants in `ALL_CAPS`. Python won't prevent reassignment — it's a signal to other developers:

````python
PI = 3.14159265358979
MAX_CONNECTIONS = 100
DATABASE_URL = "postgresql://localhost/mydb"

print(PI)
print(MAX_CONNECTIONS)
````
