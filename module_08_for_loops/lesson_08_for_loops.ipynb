# Module 08: For Loops

The `for` loop iterates over any sequence (list, string, range, etc.). It's Python's most common loop.

## 8.1 Basic `for` Loop

```python
for item in sequence:
    # body runs once per item
```

````python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)

# Also works with strings (iterates character by character)
for char in "Python":
    print(char, end=" ")   # P y t h o n
print()  # newline at end
````

## 8.2 The `range()` Function

`range()` generates a sequence of integers — the standard way to loop a specific number of times:

| Call | Sequence |
|------|----------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(2, 7)` | 2, 3, 4, 5, 6 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 |
| `range(10, 0, -1)` | 10, 9, 8, ..., 1 |

````python
# Loop 5 times
for i in range(5):
    print(i, end=" ")   # 0 1 2 3 4
print()

# Start at 1
for i in range(1, 6):
    print(i, end=" ")   # 1 2 3 4 5
print()

# Step by 2
for i in range(0, 20, 3):
    print(i, end=" ")   # 0 3 6 9 12 15 18
print()

# Count backwards
for i in range(5, 0, -1):
    print(i, end=" ")   # 5 4 3 2 1
print()
````

## 8.3 `enumerate()` — Loop With Index

When you need both the index and the value, use `enumerate()`:

````python
fruits = ["apple", "banana", "cherry", "date"]

# Without enumerate (clunky):
for i in range(len(fruits)):
    print(f"{i}: {fruits[i]}")

print()

# With enumerate (clean and Pythonic):
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

# Start index at 1 instead of 0:
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")
````

## 8.4 `zip()` — Loop Over Multiple Sequences Together

````python
names  = ["Alice", "Bob", "Carol"]
scores = [92, 87, 95]
cities = ["London", "Paris", "Tokyo"]

# Pair up elements from two lists:
for name, score in zip(names, scores):
    print(f"{name}: {score}")

print()

# Three lists at once:
for name, score, city in zip(names, scores, cities):
    print(f"{name} ({city}): {score}")
````

## 8.5 `break`, `continue`, and `else` in For Loops

Same as while loops — `break` exits early, `continue` skips to next iteration, `else` runs if loop completes without `break`:

````python
# break — stop searching once found
numbers = [3, 7, 2, 9, 4, 11, 6]
for n in numbers:
    if n > 10:
        print(f"First number > 10: {n}")
        break

# continue — skip certain values
print("Even numbers in list:")
for n in numbers:
    if n % 2 != 0:
        continue
    print(n, end=" ")
print()

# else — runs only if loop did not break
for n in numbers:
    if n > 100:
        print("Found a big number")
        break
else:
    print("No numbers > 100 found")
````

## 8.6 Nested For Loops

A loop inside a loop. The inner loop runs completely for each iteration of the outer loop.

````python
# Multiplication table
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i}x{j}={i*j}", end="  ")
    print()  # newline after each row

print()

# Pattern printing
for row in range(1, 6):
    print("*" * row)
````

## 8.7 Looping Over Dictionaries

````python
person = {"name": "Alice", "age": 30, "city": "London"}

# Loop over keys (default)
for key in person:
    print(key)

# Loop over values
for value in person.values():
    print(value)

# Loop over key-value pairs (most common)
for key, value in person.items():
    print(f"  {key}: {value}")
````

## ✅ Module 08 Summary

| Function/Keyword | Purpose |
|------------------|---------|
| `for x in seq:` | Iterate over a sequence |
| `range(n)` | Generate 0 to n-1 |
| `range(a,b,step)` | Flexible range |
| `enumerate()` | Index + value |
| `zip()` | Pair up multiple sequences |
| `break` / `continue` | Exit / skip |
