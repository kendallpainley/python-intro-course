## 1.2 How Python Works

Python is an **interpreted** language — no compilation step needed:

```
Your Code (.py)
      ↓
Python Interpreter
      ↓
Bytecode (hidden .pyc)
      ↓
Python Virtual Machine
      ↓
   Output!
```

You just write code and run it. The interpreter handles the rest.

> **Python version:** This course uses **Python 3.10+**. Python 2 is obsolete — never use it.

## 1.3 Your First Program

By tradition, every programming course starts with printing `Hello, World!`.

In Python, this is exactly **one line**:

````python
print("Hello, World!")
````

## 1.4 Comments

**Comments** are notes for humans — Python completely ignores them.

Use `#` to start a comment:

````python
# This entire line is a comment
print("Hello!")  # This is an inline comment — after the code

# Comments explain WHY, not just what:
# We greet the user by name to make the app feel personal
print("Welcome, Alex!")
````

### Multi-line comments

Python has no dedicated multi-line comment syntax. Use multiple `#` lines, or a triple-quoted string (technically a string, not a comment, but commonly used):

````python
# Option 1: multiple # lines
# This block explains the program's purpose.
# It can span as many lines as you need.

"""
Option 2: triple-quoted string.
This is actually a string value — not a true comment —
but it works fine as documentation when not assigned to anything.
"""

print("Code continues here!")
````

## 1.5 Python Is Case-Sensitive

`print`, `Print`, and `PRINT` are three different things. Only `print` (lowercase) is valid:

````python
print("This works!")  # ✓
# Print("This fails!")  # ✗ NameError: name 'Print' is not defined
# PRINT("This fails!")  # ✗ same error
````

## 1.6 Indentation Is Syntax

Most languages use `{}` braces for code blocks. Python uses **indentation** (whitespace at the start of lines). Getting indentation wrong is a syntax error.

We'll see why this matters in loops and functions. For now, the rule is: **never add random spaces at the beginning of a line.**

````python
# Correct — no extra indentation at top level:
print("Line 1")
print("Line 2")

# The following would cause an IndentationError if uncommented:
#     print("Bad indentation!")  # ← extra spaces before print
````

## 1.7 The `print()` Function — Full Options

`print()` has extra parameters worth knowing:

````python
# Multiple values — joined by a space by default
print("Python", "is", "awesome")

# Change the separator with sep=
print("2024", "01", "15", sep="-")   # → 2024-01-15
print("a", "b", "c", sep=", ")      # → a, b, c

# Change the line ending with end= (default is newline)
print("Loading", end="")
print(".", end="")
print(".", end="")
print(".")  # → Loading...

# Print a blank line
print()
````
