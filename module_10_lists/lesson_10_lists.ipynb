# Module 10: Lists

Lists are Python's most versatile built-in data structure — ordered, mutable sequences that can hold any mix of types.

## 10.1 Creating Lists

```python
# Empty list
empty = []

# List of integers
numbers = [1, 2, 3, 4, 5]

# Mixed types (valid, but uncommon)
mixed = [1, "hello", 3.14, True, None]

# Nested lists
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# From an iterable
from_range  = list(range(10))        # [0,1,2,...,9]
from_string = list("Python")          # ['P','y','t','h','o','n']

print(numbers)
print(mixed)
print(from_range)
```

## 10.2 Indexing and Slicing

Same rules as strings — zero-indexed, negative indices count from end:

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

# Indexing
print(fruits[0])    # apple
print(fruits[-1])   # elderberry
print(fruits[2])    # cherry

# Slicing  [start:stop:step]
print(fruits[1:3])  # ['banana', 'cherry']
print(fruits[:3])   # ['apple', 'banana', 'cherry']
print(fruits[2:])   # ['cherry', 'date', 'elderberry']
print(fruits[::2])  # every other: ['apple', 'cherry', 'elderberry']
print(fruits[::-1]) # reversed
```

## 10.3 Mutability — Lists Can Be Changed

Unlike strings, lists are **mutable** — you can change their contents after creation:

```python
colors = ["red", "green", "blue"]

# Change an element
colors[1] = "yellow"
print(colors)   # ['red', 'yellow', 'blue']

# Replace a slice
colors[1:3] = ["purple", "orange", "pink"]
print(colors)   # ['red', 'purple', 'orange', 'pink']

# Delete an element
del colors[0]
print(colors)   # ['purple', 'orange', 'pink']
```

## 10.4 List Methods

Lists have many built-in methods for common operations:

```python
lst = [3, 1, 4, 1, 5, 9, 2, 6]

# Add elements
lst.append(7)         # add to end
print("append:", lst)

lst.insert(0, 99)     # insert at index 0
print("insert:", lst)

lst.extend([10, 11])  # add multiple items
print("extend:", lst)

# Remove elements
lst.remove(1)         # removes FIRST occurrence of value 1
print("remove:", lst)

popped = lst.pop()    # removes and returns last element
print("pop:", popped, "→ list:", lst)

popped2 = lst.pop(0)  # remove at index
print("pop(0):", popped2, "→ list:", lst)
```

```python
lst = [3, 1, 4, 1, 5, 9, 2, 6, 5]

# Search
print(lst.count(1))    # 2  (number of times 1 appears)
print(lst.count(5))    # 2
print(lst.index(9))    # 5  (first index of value 9)

# Sort and reverse (in-place — modifies original)
lst.sort()
print("sorted:", lst)

lst.sort(reverse=True)
print("sorted desc:", lst)

lst.reverse()          # reverses in place
print("reversed:", lst)

# Non-modifying alternatives
numbers = [3,1,4,1,5]
print("sorted():", sorted(numbers))  # returns new sorted list
print("original:", numbers)          # unchanged!
```

## 10.5 Common List Operations

```python
a = [1, 2, 3]
b = [4, 5, 6]

# Concatenation
print(a + b)         # [1,2,3,4,5,6]

# Repetition
print([0] * 5)       # [0,0,0,0,0]

# Length
print(len(a))        # 3

# Membership
print(3 in a)        # True
print(7 not in a)    # True

# Min, max, sum
nums = [3, 1, 4, 1, 5, 9]
print(min(nums), max(nums), sum(nums))  # 1 9 23
```

## 10.6 Nested Lists (2D Lists)

Lists within lists create 2D structures — great for grids, matrices, tables:

```python
grid = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Access element: grid[row][col]
print(grid[0][0])   # 1 (top-left)
print(grid[1][1])   # 5 (center)
print(grid[2][2])   # 9 (bottom-right)

# Iterate over 2D list
for row in grid:
    for val in row:
        print(f"{val:3}", end="")
    print()
```

## 10.7 Copying Lists

**Important gotcha:** simple assignment does NOT copy a list — it creates another reference to the same object:

```python
original = [1, 2, 3]

# BAD: alias, not a copy!
alias = original
alias.append(4)
print(original)   # [1, 2, 3, 4]  ← original was modified!

# GOOD: copy with slice, list(), or .copy()
copy1 = original[:]
copy2 = list(original)
copy3 = original.copy()

copy1.append(99)
print(original)   # NOT modified this time
print(copy1)      # [1, 2, 3, 4, 99]
```
